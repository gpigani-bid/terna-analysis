```python
import yaml
import pandas as pd
import csv
```

```python
with open("settings.yaml", "r") as file:
    settings = yaml.safe_load(file)

forti_manager_file = settings["forti_manager"]["file"]
```

```python

records = []

with open(forti_manager_file, encoding="utf-8") as f:
    reader = csv.reader(f)
    for row in reader:
        record = {}
        for field in row:
            # scarta campi vuoti
            if not field or "=" not in field:
                continue

            key, value = field.split("=", 1)

            # rimuove doppi apici esterni e doppi
            value = value.strip('"').strip()

            record[key] = value

        records.append(record)

df = pd.DataFrame(records)
```
