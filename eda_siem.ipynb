```python
import yaml
import pandas as pd
```

```python
with open("settings.yaml", "r") as file:
    settings = yaml.safe_load(file)

siem_file = settings["siem"]["file"]
siem_sheet = settings["siem"]["sheet"]
```

```python
df = pd.read_excel(siem_file, sheet_name=siem_sheet)
```

```python
for column in df.columns:
    print(f"## Column: {column} ##")
    print("Unique values:", df[column].nunique())
    # print("Missing:", df[column].isnull().sum(), "pct", df[column].isnull().sum() * 100 / len(df))
```

```python
# value counts for columns 
for column in ['Event Name', 'Log Source', 'Event Count', 
       'Low Level Category', 
       'EventName (custom)', 'Application Name (custom)',
       'ApplianceHost (custom)']:
    print(f"Column: {column}")
    print("Unique values:", df[column].nunique())
    print("Missing:", df[column].isnull().sum())
    print("Value counts and pct:")
    value_counts = df[column].value_counts()
    pct_counts = df[column].value_counts(normalize=True) * 100
    value_pct_df = pd.DataFrame({'Value Counts': value_counts, 'Percentage': pct_counts})
    print(value_pct_df)

    print("\n")
```
