```python
import yaml
import pandas as pd
```

```python
with open("settings.yaml", "r") as file:
    settings = yaml.safe_load(file)

client = settings["forescout_client"]["file"]
server = settings["forescout_server"]["file"]
wo_violations = settings["forescout_wo_violations"]["file"]
```

```python
df_client = pd.read_csv(client, sep=";")
df_server = pd.read_csv(server, sep=";") 
df_wo_violations = pd.read_csv(wo_violations, sep=";")
```

```python
print(df_client.columns)
print(df_server.columns)
print(df_wo_violations.columns)
```
