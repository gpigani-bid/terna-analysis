```python
import yaml
import pandas as pd
```

```python
with open("settings.yaml", "r") as file:
    settings = yaml.safe_load(file)

whitelisting = settings["whitelisting"]["file"]
```

```python
df_whitelisting = pd.read_csv(whitelisting, sep=";")
```

```python
df_whitelisting['Event Name'].value_counts(), df_whitelisting['Low Level Category'].value_counts()
```

```python
whitelist_dict = {}
for c in df_whitelisting.columns:
    whitelist_dict[c] = {}
    whitelist_dict[c]['unique_values'] = df_whitelisting[c].nunique()
    whitelist_dict[c]['value_counts_top5'] = df_whitelisting[c].value_counts().head(5).to_dict()
# 
```

```python
import json
with open("whitelist_dict.json", "w") as f:
    json.dump(whitelist_dict, f)
```
